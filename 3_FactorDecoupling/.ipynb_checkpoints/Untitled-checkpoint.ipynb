## HML and RMW performance

In the previous post we conculded, that, for example, the Value factor provides positive or negative returns depending on whether the market rewards Value or Growth of companies. Let us see how the HML factor performed in 2010 - 2018.

```python
# Load factors
df = pd.read_csv('Data/ff_5factors.CSV', index_col=0 ,parse_dates=True) 

#Keep relevant years
df_old = df.loc['2000':'2009']
df_new = df.loc['2010':'2018']

# Plot compound
ret_old = (1+ df_old/100).cumprod() - 1
ret_new = (1+ df_new/100).cumprod() - 1

#PLot

_ = plt.plot(ret_old['HML'], color='maroon', label = 'HML')
_ = plt.plot(ret_old['Mkt-RF'], color='black', label = 'Mkt-RF')
_ = plt.xlim('2000','2009')
_ = plt.suptitle('Historic compound performance HML vs. Mkt-RF factors', fontsize=18)
_ = plt.title('(2000 - 2009)', fontsize= 12)
_ = plt.legend()





```

```python
_ = plt.plot(ret_new['HML'], color='maroon', label = 'HML')
_ = plt.plot(ret_new['Mkt-RF'], color='black', label = 'Mkt-RF')
_ = plt.xlim('2010','2018')
_ = plt.suptitle('Historic compound performance HML vs. Mkt-RF factors', fontsize=18)
_ = plt.title('(2010 - 2018)', fontsize= 12)
_ = plt.legend()
```

Almost flat compared to booming market factor.

What about the Profitability (RMW) factor?

```python
_ = plt.plot(ret_old['RMW'], color='maroon', label = 'RMW')
_ = plt.plot(ret_old['Mkt-RF'], color='black', label = 'Mkt-RF')
_ = plt.xlim('2000','2009')
_ = plt.suptitle('Historic compound performance RMW vs. Mkt-RF factors', fontsize=18)
_ = plt.title('(2000 - 2009)', fontsize= 12)
_ = plt.legend()
```

```python
_ = plt.plot(ret_new['RMW'], color='maroon', label = 'RMW')
_ = plt.plot(ret_new['Mkt-RF'], color='black', label = 'Mkt-RF')
_ = plt.xlim('2010','2018')
_ = plt.suptitle('Historic compound performance RMW vs. Mkt-RF factors', fontsize=18)
_ = plt.title('(2010 - 2018)', fontsize= 12)
_ = plt.legend()

# Save the plot
```

## Quality Index

Finally let's look at the Deautche Bank Equity Sector-Neutral Quality Factor. You can read more about the index composition here (https://index.db.com/dbiqweb2/servlet/indexsummary?redirect=benchmarkIndexSummary&indexid=99000242&currencyreturntype=EUR-Local&rebalperiod=2&pricegroup=STD&history=4&reportingfrequency=1&returncategory=ER&indexStartDate=20150813&priceDate=20180813&isnew=true).

In brief, it is a market neutral index that is supposed to measure the quality of the companies by investing into companies whose return on invested capital (ROIC) is higher than the average for the sector, and selling short stocks of the companies whose ROIC underperforms sector average. 

The index is calculated in euro, so we convert it in the USD using the daily close exchange rates. 

First lets see how the index performed in 2000 - 2009.

```python
#Combine index and X-rate
dbq = pd.read_csv('Data/DBGLSNQE.csv', index_col=0, parse_dates=True)
eur = pd.read_csv('Data/EURUSD.csv', index_col=0, parse_dates=True)

dbq_usd = dbq.join(eur, how='left')
dbq_usd['DBGLSQE'] = dbq_usd['DBGLSQE'] / dbq_usd['EUR']

# Calculate return
dbq_usd = dbq['DBGLSQE'].pct_change()
dbq_usd = pd.DataFrame(dbq_usd)

#Add market
dbq_usd = dbq_usd.join(df['Mkt-RF']/100, how='left')
dbq_usd = dbq_usd.iloc[1:,]

# Divide into periods
dbq_old = dbq_usd.loc['2000':'2009']
dbq_new = dbq_usd.loc['2010':'2018']

# Plot compound
dbq_ret_old = (1+ dbq_old/100).cumprod() - 1
dbq_ret_new = (1+ dbq_new/100).cumprod() - 1

#PLot

_ = plt.plot(dbq_ret_old['DBGLSQE'], color='maroon', label = 'DBGLSQE (USD)')
_ = plt.plot(dbq_ret_old['Mkt-RF'], color='black', label = 'Mkt-RF')
_ = plt.xlim('2000','2009')
_ = plt.suptitle('Historic compound performance DBGLSQE vs. Mkt-RF factors', fontsize=18)
_ = plt.title('(2000 - 2009)', fontsize= 12)
_ = plt.legend()
```

Now let us look at the recent years.

```python
_ = plt.plot(dbq_ret_new['DBGLSQE'], color='maroon', label = 'DBGLSQE (USD)')
_ = plt.plot(dbq_ret_new['Mkt-RF'], color='black', label = 'Mkt-RF')
_ = plt.xlim('2010','2018')
_ = plt.suptitle('Historic compound performance DBGLSQE vs. Mkt-RF factors', fontsize=18)
_ = plt.title('(2010 - 2018)', fontsize= 12)
_ = plt.legend()
```
